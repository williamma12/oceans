```python
import pandas as pd
import numpy as np
import sqlite3
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import date2num

%matplotlib inline
```

```python
conn = sqlite3.connect('oceanDB.sqlite')
c = conn.cursor()
```

```python
temp_query = '''\
SELECT
    DATE
    , TMAX
    , TMIN
    , TAVG
  FROM data
  WHERE
    TAVG IS NOT NULL
    AND TMIN IS NOT NULL
    AND TMAX IS NOT NULL
  ORDER BY DATE DESC;
'''
temp_df = pd.read_sql(temp_query, conn)
temp_df['DATE'] = pd.to_datetime(temp_df['DATE'], format='%Y-%m')
```

```python
dates = temp_df['DATE']
X = [date2num(date) for date in dates]
tMax = temp_df['TMAX']
tMin = temp_df['TMIN']
tAvg = temp_df['TAVG']

fig, ax = plt.subplots(figsize=(15, 5))

ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

ax.plot(dates, tMax, label='Max temperature by month')
ax.plot(dates, tMin, label='Min temperature by month')
ax.plot(dates, tAvg, label='Average temperature by month')

ax.format_xdata = mdates.DateFormatter('%Y')
ax.set_title("Temperature at San Diego International Airport by Month")
ax.set_xlabel("Year")
ax.set_ylabel("Temperature in Farenheit")
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```

From the temperature plot above, we can see that the temperature data does not get significantly distorted regardless of what aggregation function we apply to it. Thus, we can safely use the average temperature data with little loss of information. 

However, there is too much noise to draw any statistically significant inferences from the data. Thus, we will average the data by each year to determine if we can obtain any useful information from the temperature data.

```python
temp_yearly_query = '''\
SELECT
  strftime('%Y', DATE) AS year
  , avg(TAVG) AS average
  FROM data
  GROUP BY 1;
  '''
temp_yearly_df = pd.read_sql(temp_yearly_query, conn)
```

```python
year = temp_yearly_df['year']
temp = temp_yearly_df['average']

fig, ax = plt.subplots(figsize=(15, 5))

ax.plot(year, temp, label='Average temperature by year')

ax.set_title("Temperature at San Diego International Airport by Year")
ax.set_xlabel("Year")
ax.set_ylabel("Temperature in Farenheit")
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```

From the plot above, we can see that until ~2013, the average temperature by year was rarely over 65 degrees farenheit. However, at 2014 and 2015, there was a huge spike up to over 68 degrees farenheit.
